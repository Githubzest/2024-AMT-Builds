```python
#use fuzzywuzzy to match marketing approved customer lists with  dsp customer lists
#importing libraries
#*****************************************
#Step 1 - Set up
#*****************************************

from fuzzywuzzy import fuzz 
import pandas as pd 
import numpy as np
import os

# list files in the directory to confirm the files are picked from the correct location
print(os.getcwd())
print (os.listdir(os.getcwd()))

# readinng csv files into dataframes

df_cust= pd.read_csv("teal-cleaned.csv", encoding = "ISO-8859-1")
df_ads= pd.read_csv("ads-cleaned.csv", encoding= "ISO-8859-1")

# print first three rows to confirm the file read contents

df_cust.head(3)
df_ads.head(3)
print (df_cust.head(3))
print (df_ads.head(3))

print('**** Get the count of records in the Raw file ')
df_cust.count()
df_ads.count()
print (df_cust.count())
print (df_ads.count())
print (df_cust.dtypes)
print (df_ads.dtypes)
print('**** End reading count of records in the Raw file ')


print("***** Step-1 Completed Successfuly")


#************************************************
#Step 2 - Data Cleaning and Preparation
#************************************************

# removing duplicates

df_cust = df_cust.drop_duplicates()
df_ads = df_ads.drop_duplicates()

df_cust.count()
df_ads.count()
print (df_cust.count())
print (df_ads.count())


# renaming columns
df_teal = df_cust.rename(columns={'Tealium_Account_Name': 'Teal_Account'})
df_ads = df_ads.rename(columns={'Ads_Account_Name': 'Ads_Account'})

# print frame data to confirm the rename did not mixup anything
print('************* Before Merge ********')
print (df_teal.head(3))
print (df_ads.head(3))
print('************* End Before  ********')


#Cross Joint and print to confirm

print('*** Fill with na to remove the float error')

df_teal = df_teal.fillna(' ')
df_ads = df_ads.fillna(' ')

print('** start cross **')

df = df_teal.merge(df_ads, how='cross')

print('************* After Merge with Cross option ********')

print (df_teal.head(3))

print('************* Last Five records After Merge with Cross Option ********')

print (df_teal.tail(5))

print('************* End After  Merge with Cross Option ********')

df_teal['Teal_Account'] = df_teal['Teal_Account'].astype('str')

df_ads['Ads_Account'] = df_ads['Ads_Account'].astype('str')


print(df_teal.dtypes)
print(df_ads.dtypes)

print('***** Step 2 Completed Sucessfuly ***')


#*************************************
#Step -3 compare the two dataframes and create a new dataframe with the similarity score
#*************************************

print('***Starting Step - 3 ***')

# Add columns similarity_score that contains match score

# df['similarity_score'] = df.apply(lambda x: fuzz.ratio(x['Teal_Account'], x['Ads_Account']), axis=1,result_type='broadcast')
# df['similarity_score'] = df.apply(lambda x: fuzz.ratio(x['Teal_Account'], x['Ads_Account']), axis=1, result_type='expand')
# df['similarity_score'] = df.apply(lambda x: fuzz.ratio(x['Teal_Account'], x['Ads_Account']), axis=1)


## New code

# df.dtypes



#for i in df_ads.Ads_Account.unique():
#    df[i] = df.apply(lambda x: fuzz.ratio(x['Teal_Account'], x['Ads_Account']), axis=1, result_type='expand')
#    similarity_score = df[i].to_frame()
#    similarity_score.rename(columns={i: 'similarity_score'}, inplace=True)
#    df = pd.concat([df, similarity_score], axis=1)   
#    print(df.head(3))   

## end New

df['similarity_score'] = df.apply(lambda x: fuzz.ratio(x['Ads_Account'], x['Teal_Account']), axis=1)

#print(df['similarity_score'])

#filter to similarity_score above 70%

result_df = df[df['similarity_score'] > 70]

print(result_df.head(3))

print('**** Step - 3 Completed successfully ****')

#**************************
#Step -4 Import to CSV file
#**************************

#export to csv

result_df.to_csv('merged_match_rate_above_70.csv')

print('**** Step - 4 Completed successfully ****')
```
